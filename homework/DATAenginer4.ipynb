```python
import pandas as pd
import numpy as np

data = pd.read_csv('data/President_heights.csv.csv')

print(data.head())
```

```python
heights = np.array(data['height(cm)'])
print(heights)
```

```python
mean_height = np.mean(heights)
std_dev = np.std(heights)
min_height = np.min(heights)
max_height = np.max(heights)
percentiles = np.percentile(heights, [25, 50, 75])

print(f"Mean height = {mean_height}")
print(f"Standard deviation = {std_dev}")
print(f"Minimum height = {min_height}")
print(f"Maximum height = {max_height}")
print(f"25th percentile = {percentiles[0]}")
print(f"Median = {percentiles[1]}")
print(f"75th percentile = {percentiles[2]}")
```

```python
max_idx = np.argmax(heights)
min_idx = np.argmin(heights)
print(max_idx)
print(min_idx)
```

```python
max_name = data.iloc[max_idx]['name']
min_name = data.iloc[min_idx]['name']

print(f"The tallest president is {max_name}")
print(f"The smallest president is {min_name}")
```

```python
from matplotlib import pyplot as plt
import seaborn
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
seaborn.set()

plt.hist(heights, bins=10)
plt.title('Height Distribution of US President')
plt.xlabel('Height (cm)')
plt.ylabel('Number')
plt.show()
```
